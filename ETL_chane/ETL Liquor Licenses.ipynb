```python
#import tools necessary to complete project
import pandas as pd
from sqlalchemy import create_engine
import pymongo
import numpy as np
```

```python
#read CSV file and bring into pandas
liquor_csv = "chicago_liquor_licenses1.csv"
liquor_license_df = pd.read_csv(liquor_csv)
liquor_license_df.head()
```

```python
#rename rows to lower case
liquor_license_df = liquor_license_df.rename(
    columns=str.lower
)
liquor_license_df.head()
```

```python
#remove rows of locations that are not in the city of Chicago
liquor_license_df = liquor_license_df[liquor_license_df['city']=='CHICAGO']
liquor_license_df.head()
```

```python
#create conncetion to MongoDB
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
#create restaurantsDB mongo database
#create the liquor license collection
db = client.restaurantsDB
collection = db.liquor
```

```python
#remove NaN and filling with empty string
liquor_license_df = liquor_license_df.fillna('')
liquor_license_df.head()
```

```python
#put liquor_license_df in list of dictionaries then insert list of dictionaries into collection
db.liquor.insert_many(liquor_license_df.to_dict('records'))
```

```python
#print each liquor license and corresponding data across row
liquor_licenses = db.liquor.find()
for liquor_license in liquor_licenses:
    print(liquor_license)
```
