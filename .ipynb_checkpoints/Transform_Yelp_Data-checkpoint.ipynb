```python
import pandas as pd
import numpy as np
```

```python
yelp_df=pd.read_csv('Extraced_Yelp_Data.csv')
```

```python
price=[]
for i in np.arange(len(yelp_df['price'])):
    price.append(yelp_df.iloc[i]['price'].count('$'))
```

```python
yelp_df['price']=price
```
